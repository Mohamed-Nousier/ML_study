```python

# %%time

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error


# By setting it to a fixed number (like 0), 
# you ensure that every time you run this code, 
# you get the exact same sequence of "random" numbers. This is crucial for debugging machine learning models.



np.random.seed(0)
X = 2 * np.random.rand(100,1) #(row, columns)
y = 4 + 3*X + np.random.randn(100,1)
```

```python
X[:10]
```

```python
y[:10]
```

```python

# Split the data into training set and test set 
X_train, X_test, y_train, y_test =  train_test_split(
    X,
    y,
    test_size=0.2,
    random_state=0,
)
```

```python
X_train.shape, y_train.shape, y_train.shape, y_test.shape
```

```python
# Create linear-regression model
model = LinearRegression()

# Fit the model to the training data
model.fit(X_train, y_train)

# Make prediction on the testing data
y_pred = model.predict(X_test)
```

```python
y_pred[:10], y_test[:10]
```

```python
# Calculate the mean-square error
mse = mean_squared_error(y_test,y_pred)
print(f"Means Squared Error is : {mse}")
```

```python
plt.figure(figsize=(8, 4))


plt.scatter(X_train, y_train, color='blue', label='Training data')
plt.scatter(X_test, y_test, color='red', label='Testing data')
plt.plot(X_test,y_pred,color='green',linewidth=2, label='Regression Line')

plt.xlabel("Independent Variable (X)")
plt.ylabel("Dependent Variable (y)")
plt.title("Linear Regression: Fit vs Actual")

plt.grid(True, linestyle="--", alpha=0.5)


plt.legend()
plt.show()
```
